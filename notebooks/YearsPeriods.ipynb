```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier
plt.rcParams['figure.figsize'] = (18, 5)
plt.rcParams['font.family'] = 'sans-serif'
```

```python
from genre2period import get_time_period

#column names
cnames = ['year', 'genre', 'title', 'authors']

data = pd.read_csv('/home/jvdzwaan/data/embem/corpus/corpus_big.csv', sep='\t', index_col=0, header=None, names=cnames)
```

# Number of texts per period

(Corpus big+ceneton)

```python
# add period information
data['period'] = [get_time_period(y) for y in data['year']]

counts_per_period = data['period'].value_counts()

order = ['renaissance', 'classicism', 'enlightenment']

for p in order:
    print p, '\t', counts_per_period[p]
```

```python
pd.DataFrame.from_dict({'period': order, 'count': [counts_per_period[c] for c in order]}).plot(kind='bar', x='period')
```

# Number of texts per year

For all years

(Corpus big+Ceneton)

```python
# counts per year
counts_per_year = data['year'].value_counts()
years = range(1600, 1831)
counts = [counts_per_year.get(y, 0) for y in years]
count_data = pd.DataFrame.from_dict({'count': counts, 'year': years})
count_data.plot(kind='bar', x='year', y='count')
```
